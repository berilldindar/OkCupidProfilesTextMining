```python
## Veri analizi için pandas ve numpy kullanıyoruz.
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('okcupid_profiles.csv')
```

```python
## Sayfamızın ilk 10 sütununa baktığımızda dataframeleri karşılaştırıyoruz.
df.iloc[:,0:10].head()
```

```python
# Etnik kökenler için veri bütünlüğü sağlıyoruz
## Son olarak, gelir durumu için '-1'i değiştirmemiz gerekiyor
## "NaN'ler" ile değerler.
df.iloc[:,10:20].head()
```

```python
#Sadece temizlememiz gereken kısmı alıyoruz
df.iloc[:,20:].head()
```

```python
#etnik köken, boy ve gelir konusundaki yanıtları karşılaştırıyoruz.
space_camp = df[df['education'].str.contains("space camp")==True]
space_camp = space_camp[['education', 'ethnicity', 'height', 'income']]
space_camp.head(20)
```

```python
## Alanla ilişkili kullanıcılara bakarız. 
## gelir durumu $262607,76 olarak çıkıyor
space_camp_income = space_camp['income']
space_camp_income = space_camp[space_camp_income != -1]
print("Average income for users associated with space camp:",
      ('%.2f' % space_camp_income['income'].mean()))
```

```python
## Üniversitenin gelirini kontrol edelim
## mezunlarla karşılaştırmak için

masters_income = df[df['education'].str.contains('graduated from college/university')==True]
masters_income = masters_income[masters_income != -1]

print("Average income for users with masters:",
      ('%.2f' % masters_income['income'].mean()))
```

```python
all_income = df[['income']]
all_income.loc[all_income.income != -1].mean()
```

```python
## İlişkili kullanıcılar için maaş verilerinin kaldırılması
## uzay kampı ve '-1' değerlerinin değiştirilmesi  "NaN'ler" değerler ile

df.income = df['income'].mask(df.education.str.contains('space camp', na=False))
df[['education', 'income']].loc[df.education.str.contains('space camp', na=False)].head()
```

```python
## HTML sonlarını ve HTML satır sonlarını temizleme
## boşluklarla değiştirerek temizleriz

df['essay0'] = df['essay0'].str.replace('<br />', ' ')
df['essay0'] = df['essay0'].str.replace('\n', ' ')
df['essay1'] = df['essay1'].str.replace('<br />', ' ')
df['essay1'] = df['essay1'].str.replace('\n', ' ')
df['essay2'] = df['essay2'].str.replace('<br />', ' ')
df['essay2'] = df['essay2'].str.replace('\n', ' ')
df['essay3'] = df['essay3'].str.replace('<br />', ' ')
df['essay3'] = df['essay3'].str.replace('\n', ' ')
df['essay4'] = df['essay4'].str.replace('<br />', ' ')
df['essay4'] = df['essay4'].str.replace('\n', ' ')
df['essay5'] = df['essay5'].str.replace('<br />', ' ')
df['essay5'] = df['essay5'].str.replace('\n', ' ')
df['essay6'] = df['essay6'].str.replace('<br />', ' ')
df['essay6'] = df['essay6'].str.replace('\n', ' ')
df['essay7'] = df['essay7'].str.replace('<br />', ' ')
df['essay7'] = df['essay7'].str.replace('\n', ' ')
df['essay8'] = df['essay8'].str.replace('<br />', ' ')
df['essay8'] = df['essay8'].str.replace('\n', ' ')
df['essay9'] = df['essay9'].str.replace('<br />', ' ')
df['essay9'] = df['essay9'].str.replace('\n', ' ')

df[['essay0']].iloc[0,0]
```

```python
## Ayrı olan kullanıcıların boy uzunluklarını değiştirme toplamlarından 5inin ortalaması
## cdc.gov'a göre, ortalama boy uzunluğu kadınlar ve erkekler için 63.8 ve 69.3 inç
## ortalama standart boy uzunluğu 66,55 inç.
## sapma 3,00 inçtir. Böylece, kullanacağımız 51,55 ila 81,55 inç arasındaki boy uzunluğu ile değiştiririz
## geri kalanı "NaN's" ile değiştiririz

df['height']=df['height'].mask(df.height <= 51.55)
df['height']=df['height'].mask(df.height >= 81.55)
```

```python
df[['ethnicity', 'income', 'height']].loc[df['ethnicity'].str.len() >= 40].head(25)
```

```python
## Kullanıcıların ortalama gelirini bulmak için
## 3'ten fazla etnik köken

## Gelirleri diğerlerine kıyasla iki katından fazla
## Önceki mantığımızı takip ederek, uzunluktan fazlasına sahip olanlar için geliri kaldırın
## 3 etnik köken

df[['ethnicity', 'income', 'height']].loc[df['ethnicity'].str.len() >= 40].mean()
```

```python
## Kullanıcı gelirini 3'ten fazla ile değiştirme
## "NaN's" için etnik köken

df['income']=df['income'].mask(df.ethnicity.str.len() >= 40)
df[['ethnicity','income']].loc[df['ethnicity'].str.len() >= 40].head()
```

```python
## şimdi kullanıcılar üzerinde görselleştirme yapalım
## Cinsiyete göre boy için 3 karşılaştırma yaparız

## cdc.gov'dan alınan ABD verilerine göre, ortalama boy uzunluğu kadın için 63.8 inç ve erkek için 69.3 inç
df[['ethnicity', 'sex', 'height']].loc[(df.ethnicity.str.len() >= 40)].groupby('sex').mean().reset_index()
```

```python
## tırnak içinde biçimlendirme hatalarını temizleriz
df[['offspring']].head()
```

```python
## HTML alıntılarını alıntı ile değiştirerek temizleriz
df['offspring'] = df['offspring'].str.replace('&rsquo;', '\'')
df[['offspring']].head()
```
