```python
## Yaş değişkeni için 59946 veri olduğunu görüyoruz. Verisetimizde age değişkenine ait değerlerin minimum ve maksimum hangi değerleri alabildiğini
#yüzdelik olarak ortalama kaç yaşında olan insanların bu uygulamayı kullandığını gözlemleyebiliriz.
age = df[['age']]
age.describe()
```

```python
#body_type değişkeni ve cinsiyet değişkenini birleştirdik ve toplamda body_type değişkeni için insanların fiziksel olarak nasıl gözüktüğünü inceledik
body_type = df[['body_type', 'sex']]
body_type_grouped = body_type.groupby('body_type')
body_type_grouped = body_type_grouped.count().reset_index()
body_type_grouped.rename(index=str, columns={"sex":"count"})
```

```python
#Uygulamaya kayıt olan insanların beslenme şekillerini inceledik
diet = df[['diet', 'sex']]
diet_grouped = diet.groupby('diet')
diet_grouped = diet_grouped.count()
diet_grouped = diet_grouped.reset_index()
diet_grouped.rename(index=str, columns={"sex":"count"})
```

```python
## Verisetinde eğiyim seviyesine ait değişkeni inceledik ve eğitim seviyesini inceledik
education = df[['education', 'sex']]
education_grouped = education.groupby('education')
education_grouped = education_grouped.count().reset_index()
education_grouped.rename(index=str, columns={"sex":"count"})
```

```python
## Etnik kökenlere bir göz atalım 
ethnicity = df[['ethnicity', 'sex']]
ethnicity_grouped = ethnicity.groupby('ethnicity')
ethnicity_grouped = ethnicity_grouped.count().reset_index()
ethnicity_grouped.rename(index=str, columns={"sex":"count"})
ethnicity_grouped.head(10)
```

```python
## Boy uzunluğu inç cinsinden kaydedilir. Kadınlar ortalama 65 inç yani yaklaşık olarak 1.65m
#Boy uzunluğu erkekler içinse yaklaşık olarak 1.77 m 
height = df[['height', 'sex']]
height_grouped = height.groupby('sex')
height_grouped.mean().reset_index()
```

```python
## Burada, 4 inç boyunda duran bir dişimiz var.
## Ayrıca, 1 inç boyunda bir erkek.
height = df[['height', 'sex']]
height_grouped = height.groupby('sex')
height_grouped.min().reset_index()
```

```python
## Burada 95 inç boyunda duran bir kadın var.
## Ayrıca, 95 inç boyunda bir erkek.
## Sadece referans olması için, ortalama bir NBA basketbol oyuncusu
## 79 inç boyunda...
#Kısacası bunların hatalı veriler olduğunun farkına varabiliriz

height = df[['height', 'sex']]
height_grouped = height.groupby('sex')
height_grouped.max().reset_index()
```

```python
## Boy uzunluğunun standart sapmasını alıyoruz.
height_grouped.std()
```

```python
## Gelir durumu için '-1' yanıtı şu anlama gelir: "NaN's" olarak değiştirin, böylece faktör olmaz eksi değerler hata almamıza neden olacağından onları Nan olarak değiştirdik
## Bu uç değerleri çapraz kontrol etmeliyiz.
## uç değerlere sahip diğer değişkenler de bulunmaktadır.

income = df[['income', 'sex']]
income_grouped = income.groupby('income')
income_grouped = income_grouped.count().reset_index()
income_grouped.rename(index=str, columns={"sex":"count"})
```
